```python
#Downloading data from web page: https://en.tutiempo.net/climate/ws-432950.html (For Bangalore)
#as html format

def retrive_html():
    for year in range(2013, 2019):
        for month in range(1, 13):
            if(month < 10):
                url='http://en.tutiempo.net/climate/0{}-{}/ws-432950.html'.format(month,year)                                                              
            else:
                url='http://en.tutiempo.net/climate/{}-{}/ws-432950.html'.format(month,year)
                                                                          
            texts = requests.get(url)
            text_utf = texts.text.encode('utf = 8') #utf-8 encoding
        
            #Creating path if it doesn't exists
            if not os.path.exists('Data/Html_Data/{}'.format(year)):
                os.makedirs('Data/Html_Data/{}'.format(year))
            with open('Data/Html_Data/{}/{}.html'.format(year, month), 'wb') as output:
                output.write(text_utf)

        sys.stdout.flush()
        
#Main function : for calling :retrive_html
if __name__ == '__main__':
    start_time = time.time() #To Calculate the time taken to retrive data (Start time)
    retrive_html() #Function call
    stop_time = time.time() #End time
    print("Time taken to retrieve the data {}".format(stop_time - start_time))
```

## We got data hourly basis ..so we need to average of 24 hrs to find average air index of a day

```python
def avg_data(year):
    temp_i = 0
    average = []
    for rows in pd.read_csv('Data/AQI/aqi{}.csv'.format(year), chunksize = 24): #chunksize = 24 bcz 24 hrs each day
        add_var = 0
        avg = 0.0
        data = []
        df = pd.DataFrame(data = rows)
        for index, row in df.iterrows():
            data.append(row['PM2.5'])
        for i in data:
            if type(i) is float or type(i) is int:
                add_var = add_var + i
            elif type(i) is str:
                if i!='NoData' and i!='PwrFail' and i!='---' and i!= 'InVld':
                    temp = float(i) #Typecasting if there are any integres as str Ex: '22'
                    add_var = add_var + temp
                    
        avg = add_var/24
        temp_i = temp_i+1
        
        average.append(avg)
        
    return average

#Main function : for calling : avg_data
if __name__ == '__main__':
    #lst = ""
    for inpyear in range(2013, 2019):
        lst = 'lst'+str(inpyear)
        print(lst)
        lst = avg_data(inpyear)  #lst2013 ..Calling a function : avg_data
        plt.plot(range(0,365),lst2013,label="2013 data")
        plt.plot(range(0,364),lst2014,label="2014 data")
        plt.plot(range(0,365),lst2015,label="2015 data")
        plt.plot(range(0,121),lst2016,label="2016 data")
        plt.xlabel('Day')
        plt.ylabel('PM 2.5')
        plt.legend(loc='upper right')
        plt.show()
```
